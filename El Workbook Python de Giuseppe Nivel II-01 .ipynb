## **Problema 1**
Se tiene N temperaturas. Se desea calcular su media y determinar entre todas ellas cuantas son superiores o iguales a esa media.**

````python
print("-------------------------------------------------------")
print("MEDIA DE TEMPERATURAS.") 
print("-------------------------------------------------------")

#Inicializar 
Suma = 0
Media = 0.0 
C = 0       #Contador de temperaturas que superan la media
Temp = []   #Lista vacía para almacenar temperaturas

#Entradas
while True:
    N = int(input("Ingrese cantidad de Temperaturas: "))
    if 1 <= N <= 20:
        break
    else:
        print("El número de temperaturas debe entre 1 y 20")

#Proceso
for i in range(N):
    temperatura = float(input("Ingrese Temperatura {0}: ".format(i + 1)))
    Temp.append(temperatura)
    Suma += Temp[i]

Media = Suma / N 

#Contar cuantas temperaturas son mayor que la MEDIA

for tempElement in Temp:
    if tempElement >= Media:
        C += 1
        print(f"{tempElement:8.2f}")
     
#Salida
print("\nSALIDA: ") 
print("-------------------------------------------------------") 
print (f"La media es {Media:34.2f}")
print (f"Total de temperaturas >= a la media es {C:7.2f}")
````

## **Problema 4**

Un centro numérico es un número que separa una lista de números enteros (comenzando en 1) en dos grupos de números, cuyas sumas son iguales. El primer centro numérico es el 6, el cual separa la lista (1 a 8) en los grupos: (1, 2, 3, 4, 5) y (7, 8) cuyas sumas son ambas iguales a 15. El segundo centro numérico es el 35, el cual separa la lista (1 a 49) en los grupos: (1 a 34) y (36 a 49) cuyas sumas son ambas iguales a 595. 

Desarrollar una programa en Python que reciba como entrada un número entero “n” y se buscará centros numéricos entre 1 y “n”. Considera que ese número “n” tiene que ser mayor a cero y menor a 100000. 

El programa reportará el valor de “n” así como los centros numéricos que se encuentren y también la suma de los dígitos precedentes al centro numérico. Si para un valor “n” no se encuentra ningún centro numérico el resultado mostrará el texto “No hay centro numérico”

Para hallar el centro numérico considere lo siguiente:
Si “n” es el número ingresado y “m” el centro numérico, se cumple que:

>## **2m^2= n^2+n**

Luego dado un valor de “n” se puede ir calculando “m” para cada valor desde 1 hasta “n” y si “m” es entero, será el “Centro numérico”. 


````python
# Hallar los centros numéricos dado un número "n"

def centroNumerico(num):   # Función que halla un centro numérico
    m = ((num**2 + num) / 2) ** 0.5 # esto se obtine de la fórmula, hasta aquí 
                                    # "m" es float
    x = m % 1   # prueba si es entero, entonces "m" es centro numérico
    if x == 0:
        s = 0
        m = int(m)         # para usarse en el range, tiene que ser entero
        for k in range(1, m):  # obtiene la suma de los precedentes
            s += k
        return (m, s, True)    # retorna una tupla que incluye True o False
    else:
        return (0, 0, False)

# Programa principal
centroNumDicc = {}   # dicionario que guarda centro numérico y suma de previos
hayCenNum = False
while True:
    n = int(input("Ingrese el valor de 'n' "))
    if 0 < n <= 100000:
        break
    else:
        print("'n' fuera de rango, vuelva a ingresar el valor")

for j in range(2, n+1):  # no enviá el 1, porque da un falso positivo
    m, s, res = centroNumerico(j)   # invoca a la función y desempaqueta la 
                                    # la tupla que retorna
    if res:                         # Si hay centro numérico
        centroNumDicc[m] = s        # adiciona al diccionario llave y valor
        hayCenNum = True            # si pasa por aquí hay centro numérico

if hayCenNum:
    print(f"\nNúmero ingresado fue {n}")
    print("Centros numéricos  | Suma de números previos")
    print("--------------------------------------------")
    for x, y in centroNumDicc.items():
        print(f"{x:10d} {8*' '}| {y:15d}")
else:
    print("No hay centro numérico")
````

```
     3	 |          25 - 50      | 2.865 |	  1.293    |
     4	 |          50 a más     | 4.858 |  	2.193    |
```

Los precios indicados son sin IGV (18%). Las tarifas son escalonadas y se aplica siempre la tarifa del rango correspondiente y la del rango inmediatamente anterior, ejemplo:

* Para un consumo de 28 m3, se aplica la tarifa 1 para los primeros 10 m3, la tarifa2 para los 15 m3 siguientes y la tarifa 3 para los 3 m3 restantes
Solicite el nombre el nombre y el apellido del usuario por separado, el número de suministro de 5 dígitos y el consumo en m3 (máximo 1000 m3)


````python
# Pago de SEDAPAL

def calculaPago(consumo):

    if consumo <= tarifa[0][0]:
        pagoAgua = consumo * tarifa[0][1]
        pagoAlca = consumo * tarifa[0][2]

    elif consumo <= tarifa[1][0]:
        pagoAgua = (tarifa[0][0] * tarifa[0][1]
                 + (consumo - tarifa[0][0]) * tarifa[1][1])
        
        pagoAlca = (tarifa[0][0] * tarifa[0][2]
                 + (consumo - tarifa[0][0]) * tarifa[1][2])

    elif consumo <= tarifa[2][0]:
        pagoAgua = (tarifa[0][0] * tarifa[0][1]
                 + (tarifa[1][0] - tarifa[0][0]) * tarifa[1][1]
                 + (consumo - tarifa[1][0]) * tarifa[2][1])
        
        pagoAlca = (tarifa[0][0] * tarifa[0][2]
                 + (tarifa[1][0] -  tarifa[0][0])* tarifa[1][2]
                 + (consumo - tarifa[1][0]) * tarifa[2][2])

    else:
        pagoAgua = (tarifa[0][0] * tarifa[0][1]
                 + (tarifa[1][0] - tarifa[0][0]) * tarifa[1][1]
                 + (tarifa[2][0] - tarifa[1][0]) * tarifa[2][1]
                 + (consumo - tarifa[2][0]) * tarifa[3][1])
        
        pagoAlca = (tarifa[0][0] * tarifa[0][2]
                 + (tarifa[1][0] - tarifa[0][0]) * tarifa[1][2]
                 + (tarifa[2][0] - tarifa[1][0]) * tarifa[2][2]
                 + (consumo - tarifa[2][0]) * tarifa[3][2])

    return round(pagoAgua, 2), round(pagoAlca, 2)


def datosCli():
    while True:
        nCli = input("Ingresar nombre del cliente ")
        if nCli.strip() != "" and not any(map(str.isdigit, nCli)):
            break
        else:
            print("Nombre no puede estar en blanco o contener números")

    while True:
        aCli = input("Ingresar apellido del cliente ")
        if aCli.strip() != "" and not any(map(str.isdigit, aCli)):
            break    
        else:
            print("Apellido no puede estar en blanco o contener números")

    while True:
        sCli = input("Ingresar suministro (5 dígitos) ")
        if sCli.isdigit() and len(sCli) == 5:
            break
        else:
            print("Suministro incorrecto, debe ser 5 dígitos")

    while True:
        try:
            consCli = float(input("Ingresar consumo "))
            if 0 < consCli <= 1000:
                break
            else:
                print("Consumo fuera de rango [0, 1000]")
        except ValueError:
            print("Solo debe ingresar números reales")

    return nCli, aCli, sCli, consCli


def repFactura():
    print(f"\n{15*'-'} Facturación de consumo {15*'-'}")
    print(f"Nombres y apellidos {9*' '}: {nomCli} {apeCli}")
    print(f"Número de suministro {8*' '}: {sumiCli}")
    print(f"Consumo (en metros cúbicos) {' '}: {consumoCli:7.2f}")
    print(f"\n{10*'-'} Detalles a pagar {10*'-'}")
    print(f"Cargo fijo {18*' '}: {round(cargoFijo[0],2):7.2f}")
    print("Cargo por volumen")
    print(f"\t- Agua {14*' '}: {pagoAgua:7.2f}")
    print(f"\t- Alcantarillado {4*' '}: {pagoAlca:7.2f}")
    print(f"IGV {25*' '}: {igv:7.2f}")
    print(f"{30*' '}{9*'-'}")
    print(f"Total a pagar {15*' '}: {totalPagar:7.2f}")

# Programa Principal
cargoFijo = (4.886,)
tarifa = ((10, 1.116, 0.504),
          (25, 1.295, 0.586),
          (50, 2.865, 1.293),
          (10000, 4.858, 2.193)
)

nomCli, apeCli, sumiCli, consumoCli = datosCli()

pagoAgua, pagoAlca = calculaPago(consumoCli)

igv = round((cargoFijo[0] + pagoAgua + pagoAlca) * 0.18, 2)
totalPagar = round(cargoFijo[0] + pagoAgua + pagoAlca + igv,2)

repFactura()
````
