## ***Variables***

### **Nombres de variables.**
- Siempre inicia con una letra o un guión bajo (underscore = **_** )
- No puede iniciar con un número
- Solo caracteres alfanuméricos y el guión bajo
- Python es "case sensitive", es decir se consideran diferentes las mayúscula y las minúsculas
- Asi la variable "peso" es diferente de "Peso" por ejemplo

Hay dos convenciones al nombrar una variable en Python. Podemos usar la notación de mayúsculas y minúsculas o usar guiones bajos. 
* ***Camel case*** es la práctica de escribir palabras compuestas con mayúsculas y minúsculas (por ejemplo. thisIsAVariableName).
* ***guiones bajos (_)*** para separar las palabras. Si lo prefiere, puede nombrar sus variables así: this_is_a_variable_name.

### **Declaración de variables**
En Python no hay necesidad de delarar variables. Se crean en el momento que se les asigna un valor

### **Diferencia entre variables en Python y otros lenguajes**

"La forma en la que Python maneja los objetos (se evita el nombre variables, ya se verá por qué) es un tanto diferente a lo que generalmente se está acostumbrado por otros lenguajes, y puede prestar a la confusión para quienes lo ignoren. 
En muchos otros lenguajes, como C o Visual Basic, por poner algunos ejemplos, asignar a una variable es similar a poner un valor en una caja.

## **a = 2**

Ahora la caja **a** contiene el entero 2. Entonces, el código anterior se leería «poner el valor 2 en el contenedor a«.

Asignar una variable a otra hace una copia del valor que contiene la primera y lo pone en la nueva caja.

## **b = a**

**b** es una segunda caja, con una copia del número 2. La caja a tiene una copia separada. Por ende la expresión indicaría **«poner el valor que contiene a en b, sin alterar a»**

#### **a no es b**
son dos contenedores diferentes simplemente con el mismo número.

Python, en cambio, tiene **«nombres»**. Un **«nombre» o «identificador»** es como una etiqueta atada a un objeto (en este caso un número entero).

#### **a = 1**

image.png

Aquí, el número entero **1** (que, como todas las cosas en Python, es un objeto) tiene una etiqueta llamada **a**

Si reasignamos al identificador **a**, simplemente movemos la etiqueta a otro objeto.

image.png

Ahora el nombre **a** está atado al objeto **2** (que también es un número entero).

El objeto original (el número 1) ya no tiene la etiqueta **a**. Puede que todavía exista, pero no podemos acceder a él a través de esta etiqueta. (Cuando un objeto no tiene más referencias o etiquetas, es eliminado).

Si asignamos un nombre (o identificador) a otro, solo estamos atando otra etiqueta a un objeto existente.

#### **b=a**

image.png

El nombre **b** es simplemente una segunda etiqueta atada al mismo objeto que **a**

Si bien comunmente nos referimos a **«variables»** incluso en Python (porque es una terminología convencional), en realidad queremos decir **«nombres» o «identificadores»**. En Python, las **«variables»** no son cajas que contienen valores sino **etiquetas** asociadas a un valor (objeto) determinado."

## **La función print, opciones**

### ***sep***
Hay un argumento opcional llamado **sep**, abreviatura de separador, que puede usar para cambiar ese espacio a algún otra cosa. Por ejemplo, usando sep=':' separaría los argumentos por dos puntos y sep='##' separaría los argumentos con dos signos de #. 
Una posibilidad particularmente útil es no tener nada dentro de las comillas, como en sep=””. Esto dice que no hay que separar los argumentos. Aquí hay un ejemplo en el que sep es útil para que la salida se vea bien:

```python
print('El valor de 3+4 is', 3+4, '.')
print('El valor de 3+4 is ', 3+4, '.', sep='')
```

### ***end***
La función de impresión avanzará automáticamente a la siguiente línea. Por ejemplo, lo siguiente imprimirá en dos líneas:

```python
print('En la primera línea')
print('En la segunda línea')
```

El argumento opcional llamado **end**, evita que la función de impresión avance pasando a la siguiente línea. Aquí hay un ejemplo:

```python
print('En la primera línea', end='')
print('En la segunda línea')
```

```python
print('En la primera línea', end=' ')
print('En la segunda línea')
```
