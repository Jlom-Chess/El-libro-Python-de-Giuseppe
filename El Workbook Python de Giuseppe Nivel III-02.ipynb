Escriba un programa que simule el juego Piedra = 1, Papel = 2, Tijera = 3 con las siguientes características:

•	El usuario juega contra la computadora, quien genera de manera aleatoria su jugada

•	El usuario ingresa su jugada digitando 1, 2 o 3 a través del teclado

•	Cada juego consta de tres “tiros”

•	Gana el juego el que hace más puntos en los tres tiros. Si alguno consigue dos puntos antes del tercer tiro, ya no se realiza este último tiro

•	El programa dirá quien ganó el tiro o si hubo empate, señalando lo que saco tanto el usuario como el programa

•	Concluido el juego mostrará el resultado indicando el ganador, cuantos puntos ganó cada uno y cuantos tiros empatados si hubo

•	Al inicio del programa se pedirá el nombre del usuario y el programa le dará la bienvenida al juego

•	Cuando termina el juego, el programa preguntará si quiere realizar otro juego, si contesta que sí, se inicia otro juego de tres tiros

•	Si contesta que no el programa se despedirá y mostrará un resumen de los juegos realizados mostrando un reporte similar al siguiente

   Juego   | Puntos-Usr | Puntos-Pc   |  Empates   |  Ganador 
---------------------------------------------------------------------------------
