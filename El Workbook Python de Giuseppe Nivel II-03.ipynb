```python
# Pregunta PC4 - Caso enviado por Norman

class Vehiculo():
    def __init__(self, color, ruedas):
        self.color = color
        self.ruedas = ruedas
        
    def __str__(self):
        return "color {}, {} ruedas".format(self.color, self.ruedas)
    
    
class Coche(Vehiculo):
    def __init__(self, color, ruedas, velocidad, cilindrada):
        super().__init__(color, ruedas)
        self.velocidad = velocidad
        self.cilindrada = cilindrada
        
    def __str__(self):
        return super().__str__() + ", {} Km/h, {} cc".format(
            self.velocidad, self.cilindrada)
    
    
class Camioneta(Coche):
    def __init__(self, color, ruedas, velocidad, cilindrada, carga):
        super().__init__(color, ruedas, velocidad, cilindrada)
        self.carga = carga
        
    def __str__(self):
        return super().__str__() + ", {} Kg de carga".format(self.carga)
    
    
class Bicicleta(Vehiculo):
    def __init__(self, color, ruedas, tipo):
        super().__init__(color, ruedas)
        self.tipo = tipo
    
    def __str__(self):
        return super().__str__() + ", {} ".format(self.tipo)
    
    
class Motocicleta(Bicicleta):
    def __init__(self, color, ruedas, tipo, velocidad, cilindrada):
        super().__init__(color, ruedas, tipo)
        self.velocidad = velocidad
        self.cilindrada = cilindrada
        
    def __str__(self):
        return super().__str__() + ", {} Km/h, {} cc".format(
            self.velocidad, self.cilindrada)
    

# esto es para la primera del catálaogo
#def catalogar(lista):
#    for v in lista:
#        print("{} {}".format(type(v).__name__, v))
      
# catalogar(vehiculos)

# Esta es para la segunda versión del catálogo
def catalogar(lista, ruedas=None):
    if ruedas != None:
        contador = 0
        for v in lista:
            if v.ruedas == ruedas:
                contador += 1
        print("Se han encontrado {} vehículos con {} ruedas"
              .format(contador, ruedas))
        print("==========================================")
    else:
        print("El catálogo completo es el siguiente")
        print("====================================")
        
    for v in lista:
        if ruedas == None:
            print("{} {}".format(type(v).__name__, v))
        elif v.ruedas == ruedas:
            print("{} {}".format(type(v).__name__, v))
            
# Esto es solo un ejemplo, los alumnos puede crear otros
# Además en la PC4 se dice que se ingrese dos de cada tipo
vehiculos = [
    Coche("azul", 4, 150, 1200),
    Coche("rojo", 4, 180, 1800),
    Camioneta("blanco", 4, 100, 1300, 1500),
    Camioneta("verde", 4, 100, 1400, 1600),
    Bicicleta("verde", 2, "urbano"),
    Bicicleta("azul", 2, "montañera"),
    Motocicleta("negro", 2, "deportiva", 150, 1200),
    Motocicleta("plateada", 2, "urbana", 120, 1000)
]

catalogar(vehiculos)
```

```python
class Calificación:
    def __init__(self, codigo, nombre, modalidad, pc1, pc2):
        self.codigo = codigo
        self.nombre = nombre
        self.modalidad = modalidad 
        self.pc1 = pc1
        self.pc2 = pc2
        

class CalificacionEpe(Calificación):
    def __init__(self, codigo, nombre, modalidad, pc1, pc2, tp, tf):
        super().__init__(codigo, nombre, modalidad, pc1, pc2)
        self.tp = tp
        self.tf = tf
        
    def nf(self):
        return round(0.2*self.pc1 + 0.2*self.pc2 + 0.30*self.tp + 
                     0.20*self.tf, 2)
        
    def __str__(self):
        return f"Nota final EPE: {self.nf()}"
    

class CalificacionPg(Calificación):
    def __init__(self, codigo, nombre, modalidad, pc1, pc2, pf, eb):
        super().__init__(codigo, nombre, modalidad, pc1, pc2)
        self.pf = pf
        self.eb = eb
        
    def nf(self):
        return round(0.15*self.pc1 + 0.15*self.pc2 + 0.40*self.pf + 
                     0.30*self.eb, 2)
        
    def __str__(self):
        return f"Nota final PG: {self.nf()}"    
    
      
alep = CalificacionEpe("001","Jose Ojeda", "epe", 14, 13, 16, 12)
alpg = CalificacionPg("001","Luis Merino", "pg", 14, 13, 16, 12)
print(alep)
print(alpg)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Menú principal
def menu():
    print("Sistema de Calificaciones")
    print("[1] Registrar Calificación")
    print("[2] Listar Calificaciones")
    print("[3] Mostrar Estadisticas")
    print("[4] Salir")
    opcion = ingresarOpcion("Ingrese su opción: ", 1, 4)
    return opcion


def ingresarOpcion(etiqueta, inferior, superior):
    while True:
        try:
            print(etiqueta)
            numint = int(input())
            if inferior <= numint <= superior:
                return numint
        except ValueError:
            print("Sucedió un error en el ingreso de datos, " 
                  "reintente nuevamente")


# Clase Sistema de Calificaciones
class Sistema:
    def __init__(self):
        self.dfCalificaciones = pd.read_excel("NotasAlumnos.xlsx", "Notas")

    def ingresarCalificacion(self):
        print("Ingreso de Calificación")
        print("----------------------\n")
        codigo = datosAlum.ingresarCamposCadena("Codigo Alumno: ", 1)
        if not self.existeCalificacion(codigo):
            nombre = datosAlum.ingresarCamposCadena("Nombre Alumno: ", 2)
            parcial = datosAlum.ingresarNota("Examen Parcial: ")
            final = datosAlum.ingresarNota("Examen Final: ")
            calificacion = Calificacion(codigo, nombre, parcial, final)
            self.registrarCalificacion(calificacion)
        else:
            print("Código duplicado!")

    def existeCalificacion(self, codigo):
        subDataFrame = self.dfCalificaciones.loc[
            self.dfCalificaciones["Codigo"] == codigo]
        return subDataFrame.empty is False

    def registrarCalificacion(self, calificacion):
        # Usar loc para añadir al final del DataFrame
        self.dfCalificaciones.loc[
            self.dfCalificaciones.shape[0]] = calificacion.dataRow()
        self.dfCalificaciones.to_excel("NotasAlumnos.xlsx", sheet_name="Notas",
                                       index=False)

    def listarCalificaciones(self):
        print(self.dfCalificaciones)


# Clase de los datos del alumno
class DatosAlumno:
    def __init__(self):
        self.numfloat = 0.0
        self.cadena = ""

    def ingresarNota(self, etiqueta):
        while True:
            try:
                print(etiqueta)
                self.numfloat = float(input())
                if 0 <= self.numfloat <= 20:
                    return self.numfloat
                else:
                    print("La nota debe estar en el rango [0, 20]")
            except ValueError:
                print(
                    "Sucedió un error en el ingreso de datos, reintente "
                    "nuevamente")

    def ingresarCamposCadena(self, etiqueta, tipo):
        while True:
            print(etiqueta)
            self.cadena = input()
            if self.cadena.strip() != "":
                if tipo == 1:
                    if (self.cadena[0] == "A" and len(self.cadena) == 7 and
                            self.cadena[1:].isdigit()):
                        return self.cadena
                    else:
                        print("Código debe tener la estructura: Annnnnn")
                else:
                    if not any(map(str.isdigit, self.cadena)):
                        return self.cadena
                    else:
                        print("Nombre no puede tener números")
            else:
                print("Campo obligatorio no puede estar vacío")


# Clase Entidad Calificacion
class Calificación:
    def __init__(self, codigo, nombre, modalidad, pc1, pc2):
        self.codigo = codigo
        self.nombre = nombre
        self.modalidad = modalidad 
        self.pc1 = pc1
        self.pc2 = pc2
        

class CalificacionEpe(Calificación):
    def __init__(self, codigo, nombre, modalidad, pc1, pc2, tp, tf):
        super().__init__(codigo, nombre, modalidad, pc1, pc2)
        self.tp = tp
        self.tf = tf
        
    def nf(self):
        return round(0.2*self.pc1 + 0.2*self.pc2 + 0.30*self.tp + 
                     0.20*self.tf, 2)
        
    def __str__(self):
        return f"Nota final EPE: {self.nf()}"
    
    def dataRow(self):
        return [self.codigo, self.nombre, self.modalidad, self.nf()]
    

class CalificacionPg(Calificación):
    def __init__(self, codigo, nombre, modalidad, pc1, pc2, pf, eb):
        super().__init__(codigo, nombre, modalidad, pc1, pc2)
        self.pf = pf
        self.eb = eb
        
    def nf(self):
        return round(0.15*self.pc1 + 0.15*self.pc2 + 0.40*self.pf + 
                     0.30*self.eb, 2)
        
    def __str__(self):
        return f"Nota final PG: {self.nf()}"

    def dataRow(self):
        return [self.codigo, self.nombre, self.modalidad, self.nf()]


# Programa Principal
sistema = Sistema()
datosAlum = DatosAlumno()
while True:
    opcion = menu()
    if opcion == 1:
        sistema.ingresarCalificacion()
    elif opcion == 2:
        sistema.listarCalificaciones()
    elif opcion == 3:
        sistema.mostrarEstadisticas()
    elif opcion == 4:
        break
```

```python
class Volumen():
    def __init__(self, nombre, numero):
        self.nombre = nombre
        self.numero = numero
        
    def __str__(self):
        return "Nombre del volumen: {}, numero: {}".format(self.nombre, 
                                                             self.numero)
    

class Book(Volumen):
    def __init__(self, nombre, numero, isbn):
        super(). __init__(nombre, numero)
        self.isbn = isbn
        
    def __str__(self):
        return super().__str__() + ", isbn: {}".format(self.isbn)
        
        
class NewsPaper(Volumen):
    def __init__(self, nombre, numero, issn):
        super(). __init__(nombre, numero)
        self.isbn = issn
        
    def __str__(self):
        return super().__str__() + ", issn: {}".format(self.issn)
    
    
class Library():
    def __init__(self, name, address, volumenes = []):
        self.name = name
        self.address = address
        self.volumenes = volumenes
        
    def add_volumen(self, volumen):
        self.volumenes.append(volumen)
        
    def remove_volumen(self, id):
        self.volumenes.remove(id)
        
    def __str__(self):
        return f"{self.name} {self.address} {self.volumenes[0]}"
        
        
b = Book("Prueba", 5, "as345") 
print(b)
l = Library("Lib1", "Mi casa")
l.add_volumen(b)
print(l)

    
```

```python
#11 Calendar
class Calendar:
    def __init__(self, year):
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0 or 
                (self.year % 4 == 0 and self.year % 100 != 0))

    def first_day(self, m):
        p = (14-m) // 12
        q = self.year - p
        r = q + q//4 - q//100 + q//400
        s = m + 12*p - 2
        t = (1 + r + (31*s)//12) % 7
        return t

    def print_calendar(self, m):
        month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        meses = ["Enero", "Febrero", "Marzo", "Abril", 
                 "Mayo", "Junio", "Julio", "Agosto", 
                 "Setiembre", "Octubre", "Noviembre", "Diciembre"]
        dias = ["D", "L", "M", "M", "J", "V", "S"]
        
        if self.is_leap_year() and m == 2:
            days = 29
        else:
            days = month_days[m-1]

        print(f"  {meses[m-1]}")
        print("  ", end='')
        for i in range(7):
            print(f"{dias[i]:^4s}", end='')
        print()
     
        c = 0
        for i in range(self.first_day(m)):
            c += 1
            print('    ', end='')
            
        for i in range(1, days+1):
            print('{:4d}'.format(i), end='')
            c += 1
            if c % 7 == 0:
                print()
        print()

calendar = Calendar(2021)
print(calendar.is_leap_year())
print(calendar.first_day(8))
calendar.print_calendar(8)
```
